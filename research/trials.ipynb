```python
numerical_columns = []
categorical_columns = []

for col in df.columns:
    if df[col].dtype in ['int64', 'float64']:
        numerical_columns.append(col)
    else:
        categorical_columns.append(col)

print("Numerical columns:", numerical_columns)
print("Categorical columns:", categorical_columns)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

categorical_columns = ['gender', 'ssc_board', 'hsc_board', 'hsc_subject', 
                       'undergrad_degree', 'work_experience', 'specialisation', 
                       'status', 'company_tier', 'job_competition_level']

for col in categorical_columns:
    df[col] = le.fit_transform(df[col])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Select numerical columns (excluding 'status')
numerical_columns = [col for col in df.columns if df[col].dtype in ['int64', 'float64'] and col != 'status']

# Plot boxplots
for col in numerical_columns:
    plt.figure(figsize=(8, 5))
    sns.boxplot(x='status', y=col, data=df)
    plt.title(f'Boxplot of {col} vs Status')
    plt.show()
```
